```python
#import libraries
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#load the image Data
img = cv2.imread("abc4.jpeg")
```

 **Loading Image**

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.title("Rabbit")
plt.imshow(img)
plt.show()
```

**Converting to Grayscale Image**

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.title("Gray Image")
plt.imshow(gray_img,cmap="gray")
plt.show()
```

**Inverting the Image**

```python
invert_image = cv2.bitwise_not(gray_img)
plt.title("Invert Image")
plt.imshow(invert_image, cmap="gray")
plt.show()
```

**Blurring the Inverted image**

```python
blur_img = cv2.GaussianBlur(invert_image,(111,111),0)
plt.title("Blur Image")
plt.imshow(blur_img, cmap="gray")
plt.show()
```

**Invert blur Image**

```python
invblur_img=cv2.bitwise_not(blur_img)
plt.title("Invert Blur Image")
plt.imshow(invblur_img, cmap="gray")
plt.show()
```

**Pencil Sketch Image**

```python
sketch_img=cv2.divide(gray_img, invblur_img, scale=256.0)
plt.title("Pencil Sketch Image")
plt.imshow(sketch_img, cmap='gray')
plt.show()
```
